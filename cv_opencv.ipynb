```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
resized = cv2.resize(gray,(300, 300))
cv2_imshow(resized)
```

```python
cropped = gray[198:240, 180:358]
cv2_imshow(cropped)
```

```python
rotated = cv2.rotate(gray,(cv2.ROTATE_180))
cv2_imshow(rotated)
```

```python

flipped = cv2.flip(rotated, 1)
cv2_imshow(flipped)
```
